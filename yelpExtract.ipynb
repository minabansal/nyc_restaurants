```python
import requests
import json
import pandas as pd
# Import API key
from yelp_keys import api_key
```

```python
yelpAPI = 'https://api.yelp.com/v3/businesses/search'
header = {'Authorization':'bearer %s' % api_key}
parameters = {'term':'restaurants',
             'limit':50,
             'radius':10000,
             'categories' : 'All',
             'location': 'New York City',
             'is_closed': False}

# send a request
response = requests.get(url = yelpAPI, params = parameters, headers = header)
```

```python
response
restaurant_data = response.json()
restaurant_data
```

```python
restaurantDF = pd.DataFrame()
count = 0
for i in restaurant_data['businesses']:
    temprestaurantDF = pd.DataFrame()
    temprestaurantDF.loc[count,'name'] = i['name']
    temprestaurantDF.loc[count,'address'] = i['location']['address1']
    temprestaurantDF.loc[count,'latitude'] = i['coordinates']['latitude']
    temprestaurantDF.loc[count,'longitude'] = i['coordinates']['longitude']
    temprestaurantDF.loc[count,'rating'] = i['rating']
    temprestaurantDF.loc[count,'review_count'] = i['review_count']
    temprestaurantDF.loc[count,'price'] = i['price']
    temprestaurantDF.loc[count,'display_phone'] = i['display_phone']
    restaurantDF = restaurantDF.append(pd.DataFrame(data = temprestaurantDF), ignore_index=True, sort=False)
    count = count + 1

restaurantDF
```
